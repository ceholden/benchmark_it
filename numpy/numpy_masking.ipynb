# Masking of arrays based on range of values

E.g., mask an image band based on 0 < band < 10,000 (0 - 1 reflectance)

```python
import numpy as np
```

```python
%load_ext cythonmagic
```

```python
# Test data
n_band = 7
n_mask = 50
n_images = 1000

band = np.random.randint(0, 10000, size=(n_band, n_images)).astype(np.int32)
for b in range(n_band):
    band[b, np.random.choice(np.arange(0, n_images), size=n_mask, replace=False)] = 16000
    
mins = np.repeat(0, n_band).astype(np.int16)
maxes = np.repeat(10000, n_band).astype(np.int16)
```

### Combining masks

```python
def valid_mask_numpy(band, mins, maxes):
    return np.all([((b > _min) & (b < _max)) for b, _min, _max in zip(np.rollaxis(band, 0), mins, maxes)], axis=0)
```

```python
%%cython

import numpy as np

cimport numpy as np
cimport cython

@cython.boundscheck(False)
@cython.wraparound(False)
@cython.nonecheck(False)
cdef np.ndarray[np.uint8_t, ndim=1] valid_mask(np.ndarray[np.int32_t, ndim=2] array,
                                               np.ndarray[np.int32_t, ndim=1] mins,
                                               np.ndarray[np.int32_t, ndim=1] maxes):
    cdef Py_ssize_t rows = array.shape[0]
    cdef Py_ssize_t cols = array.shape[1]
    cdef Py_ssize_t row, col
    cdef np.ndarray[np.uint8_t, ndim=1] valid = np.ones(rows, dtype=np.uint8)
    
    for row in range(rows):
        for col in range(cols):
            if array[row, col] <= mins[col] or array[row, col] >= maxes[col]:
                valid[row] = 0
                break
    return valid
```

```python
%%cython

import numpy as np

cimport numpy as np
cimport cython

@cython.boundscheck(False)
@cython.wraparound(False)
@cython.nonecheck(False)
cdef char[:] valid_mask_memview(int[:, :] array,
                                  int[:] mins, int[:] maxes):
    """ Return mask array if any values across columns are out of range
    """
    cdef Py_ssize_t rows = array.shape[0]
    cdef Py_ssize_t cols = array.shape[1]
    cdef Py_ssize_t row, col
    cdef char[:] valid = np.ones(rows, dtype=np.uint8)

    for row in range(rows):
        for col in range(cols):
            if array[row, col] <= mins[col] or array[row, col] >= maxes[col]:
                valid[row] = 0
                break
    return valid
```

```python
%%cython

import numpy as np

cimport numpy as np
cimport cython

@cython.boundscheck(False)
@cython.wraparound(False)
@cython.nonecheck(False)
cdef np.ndarray[np.uint32_t, ndim=1] valid_indices_memview(int[:, :] array,
                                          int[:] mins, int[:] maxes):
    """ Return indices of array if values fit within range
    """
    cdef Py_ssize_t rows = array.shape[0]
    cdef Py_ssize_t cols = array.shape[1]
    cdef Py_ssize_t row, col
    cdef char[:] valid_mask = np.ones(rows, dtype=np.uint8)

    cdef Py_ssize_t n_valid = rows
    cdef Py_ssize_t i_valid = 0
    
    for row in range(rows):
        for col in range(cols):
            if array[row, col] <= mins[col] or array[row, col] >= maxes[col]:
                valid_mask[row] = 0
                n_valid -= 1
                break
            
    cdef valid = np.empty(n_valid, dtype=np.uint32)
    
    for row in range(rows):
        if valid_mask[row] == 1:
            valid[i_valid] = row
            i_valid += 1
    
    return valid
```

```python
%%cython -a

import numpy as np

cimport numpy as np
cimport cython

@cython.boundscheck(False)
@cython.wraparound(False)
@cython.nonecheck(False)
cpdef np.ndarray[np.uint32_t, ndim=2] valid_data(int[:, :] array,
                                                int[:] mins, int[:] maxes):
    """ Return subset of array if values fit within range
    """
    cdef Py_ssize_t rows = array.shape[0]
    cdef Py_ssize_t cols = array.shape[1]
    cdef Py_ssize_t row, col
    cdef char[:] valid_mask = np.ones(rows, dtype=np.uint8)

    cdef Py_ssize_t n_valid = rows
    cdef Py_ssize_t i_valid = 0
    
    for row in range(rows):
        for col in range(cols):
            if array[row, col] <= mins[col] or array[row, col] >= maxes[col]:
                valid_mask[row] = 0
                n_valid -= 1
                break
                
    cdef out_array = np.empty((n_valid, cols), dtype=np.uint32)
    cdef np.uint32_t[:, :] out_view = out_array
    
    for row in range(rows):
        if valid_mask[row] == 1:
            for col in range(cols):
                out_view[i_valid, col] = array[row, col]
            i_valid += 1
    
    return out_array
```

```python
%%cython -a
import numpy as np

cimport numpy as np
cimport cython

@cython.boundscheck(False)
@cython.wraparound(False)
@cython.nonecheck(False)
cpdef np.ndarray[np.uint32_t, ndim=1] get_valid_mask(
        np.int32_t[:, :] array,
        np.int32_t[:] mins,
        np.int32_t[:] maxes):
    """ Return mask for values of array within band by band min and max ranges

    Args:
      array (np.ndarray):

    """
    cdef Py_ssize_t rows = array.shape[0]
    cdef Py_ssize_t cols = array.shape[1]
    cdef Py_ssize_t row, col
    cdef valid_mask = np.ones(rows, dtype=np.uint8)
    cdef np.uint8_t[:] valid_mask_view = valid_mask

    for row in range(rows):
        for col in range(cols):
            if array[row, col] <= mins[col] or array[row, col] >= maxes[col]:
                valid_mask_view[row] = 0
                break
                
    return valid_mask
```

```python
%%cython
import numpy as np

cimport numpy as cnp
cimport cython

ctypedef fused np_int_1:
    cnp.int16_t
    cnp.int32_t
    cnp.int64_t
    
ctypedef fused np_int_2:
    cnp.int16_t
    cnp.int32_t
    cnp.int64_t
    

@cython.boundscheck(False)
@cython.wraparound(False)
@cython.nonecheck(False)
cpdef cnp.ndarray[cnp.uint8_t, ndim=1] get_valid_mask(
        np_int_1[:, :] array,
        np_int_2[:] mins,
        np_int_2[:] maxes):
    """ Return mask for values of array within band by band min and max ranges

    Args:
      array (np.ndarray): 2D array (# observations x # band, np.int32) to mask
      mins (np.ndarray): 1D array (# band, np.int32) of minimum range bound
      maxes (np.ndarray): 1D array (# band, np.int32) of maximum range bound

    Returns:
      np.ndarray: 1D (# observations, np.uint8) array mask with 1's for
        observations falling within the range and 0's for observations outside
        the specified range

    """
    cdef Py_ssize_t rows = array.shape[0]
    cdef Py_ssize_t cols = array.shape[1]
    cdef Py_ssize_t row, col
    cdef valid_mask = np.ones(rows, dtype=np.uint8)
    cdef cnp.uint8_t[:] valid_mask_view = valid_mask

    for row in range(rows):
        for col in range(cols):
            if array[row, col] <= mins[col] or array[row, col] >= maxes[col]:
                valid_mask_view[row] = 0
                break

    return valid_mask


@cython.boundscheck(False)
@cython.wraparound(False)
@cython.nonecheck(False)
cpdef cnp.ndarray[cnp.uint8_t, ndim=1] get_valid_mask_T(
        np_int_1[:, :] array,
        np_int_2[:] mins,
        np_int_2[:] maxes):
    """ Return mask for values of array within band by band min and max ranges

    Args:
      array (np.ndarray): 2D array (# observations x # band, np.int32) to mask
      mins (np.ndarray): 1D array (# band, np.int32) of minimum range bound
      maxes (np.ndarray): 1D array (# band, np.int32) of maximum range bound

    Returns:
      np.ndarray: 1D (# observations, np.uint8) array mask with 1's for
        observations falling within the range and 0's for observations outside
        the specified range

    """
    cdef Py_ssize_t cols = array.shape[0]
    cdef Py_ssize_t rows = array.shape[1]
    cdef Py_ssize_t row, col
    cdef valid_mask = np.ones(rows, dtype=np.uint8)
    cdef cnp.uint8_t[:] valid_mask_view = valid_mask

    for row in range(rows):
        for col in range(cols):
            if array[col, row] <= mins[col] or array[col, row] >= maxes[col]:
                valid_mask_view[row] = 0
                break

    return valid_mask
```
